## Exercise 1

```python
# Some data for plotting
from matplotlib import pyplot as plt
import numpy as np
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
np_yerr = np.array(y_err)
x_err = .25
#ecolor is the color for error
#markerfacecolor is color for marker
plt.errorbar(x, y, 
             xerr=.25, yerr = np_yerr,
             color='#8B0000', linestyle='dashed',
             marker='o', markerfacecolor= 'yellow',
             ecolor='#000000',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc = 'lower right')
plt.show()
```

## Exercise 2

```python
import math
x = list(range(1,51))
x = np.linspace(0, 50, 1000)
np_x = np.array(x)
y = (math.e ** (-np_x/10))*np.sin(x)
plt.plot(x, y, color='tab:blue')
plt.xlabel('x', fontsize = 15, c = 'darkorange')
plt.ylabel('y', fontsize = 15, c = 'darkorange', rotation = 'horizontal')
plt.title(r'$y = \mathit{e}^{-x/10}\sin(x)$', fontsize = 20)
plt.grid(alpha=.25)
plt.ylim([-1,1])

#how to get rid of the (-1.0, 1.0)?
```
